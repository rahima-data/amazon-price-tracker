# 🛒 Amazon Price Tracker

A Python web scraping project that tracks Amazon product prices over time.

## 🔧 Libraries Used
- `requests` - fetches the Amazon webpage
- `BeautifulSoup` - parses the HTML
- `pandas` - reads and displays price history
- `datetime` - records the date of each price check
- `csv` - saves data to a spreadsheet

## 📊 Output
Saves price history to `amazon_price_tracker.csv` — run daily to build up a price history!

```python
# Libraries for scraping
import requests
from bs4 import BeautifulSoup

# Libraries for data and dates
import datetime
import csv
import os
import pandas as pd
```

```python
# Disguise our scraper as a real browser
header = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/147.0.0.0 Safari/537.36",
    "Accept-Language": "en-GB,en;q=0.9"
}

# Product we are tracking
url = "https://www.amazon.com/dp/B0DNY7TXQD"
```

```python
# Fetch the page
response = requests.get(url, headers=header)
soup = BeautifulSoup(response.text, "lxml")

# Extract product details
product_title = soup.find(id="productTitle").text.strip()
product_price = soup.find(class_="a-offscreen").text.strip()

# Get today's date
date = datetime.datetime.now().strftime("%Y-%m-%d")

# Print summary
print(f"📦 Product: {product_title}")
print(f"💰 Price: {product_price}")
print(f"📅 Date: {date}")
```

```python
filename = "amazon_price_tracker.csv"

# Create file with headers if it doesn't exist
if not os.path.exists(filename):
    with open(filename, "w", newline="", encoding="utf-8") as f:
        writer = csv.writer(f)
        writer.writerow(["Date", "Product", "Price"])

# Append today's data
with open(filename, "a", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow([date, product_title, product_price])

print(f"✅ Price recorded on {date}")
```
